```python
import os
from sqlalchemy import create_engine, text
from dotenv import load_dotenv
from psycopg2 import sql, connect
from pathlib import Path

def apply_migration(engine, migration_file):
    with engine.connect() as conn:
        with open(migration_file, 'r') as file:
            migration_sql = file.read()
            print(migration_sql)
            conn.execute(text(migration_sql))
            conn.execute(
                text("""
                    INSERT INTO migration_history (migration_file)
                    VALUES (:migration_file)
                """),
                {'migration_file': os.path.basename(migration_file)}
            )

def get_applied_migrations(engine):
    with engine.connect() as conn:
        result = conn.execute(text("SELECT migration_file FROM migration_history"))
        return {row['migration_file'] for row in result}

def init_db():
    load_dotenv()
    DB_USER = os.getenv("DB_USER")
    DB_PASSWORD = os.getenv('DB_PASSWORD')
    DB_HOST = os.getenv('DB_HOST')
    DB_PORT = os.getenv('DB_PORT')
    DB_NAME = os.getenv('DB_NAME')
    DB_TN_MIGRATION_HISTORY = os.getenv('DB_TN_MIGRATION_HISTORY')

    MIGRATION_DIR = Path('./../../db/migrations')

    connection_url = f'postgresql+psycopg2://{DB_USER}:{DB_PASSWORD}@{DB_HOST}:{DB_PORT}/{DB_NAME}'
    engine = create_engine(connection_url)


    # Create migration history table if not exists
    with engine.connect() as conn:
        print(conn.execute(text(f"SELECT * FROM migration_history")).keys())
        conn.execute(text(f"""
            CREATE TABLE IF NOT EXISTS {DB_TN_MIGRATION_HISTORY} (
                id SERIAL PRIMARY KEY,
                migration_file VARCHAR(255) NOT NULL,
                applied_at TIMESTAMP DEFAULT CURRENT_TIMESTAMP
            );
        """))

    applied_migrations = get_applied_migrations(engine)
    migration_files = sorted([f for f in os.listdir(MIGRATION_DIR) if f.endswith('.sql')])

    for migration_file in migration_files:
        if migration_file not in applied_migrations:
            print(f"Applying migration: {migration_file}")
            apply_migration(engine, os.path.join(MIGRATION_DIR, migration_file))
        else:
            print(f"Skipping already applied migration: {migration_file}")
```

```python
load_dotenv()
DB_USER = os.getenv("DB_USER")
DB_PASSWORD = os.getenv('DB_PASSWORD')
DB_HOST = os.getenv('DB_HOST')
DB_PORT = os.getenv('DB_PORT')
DB_NAME = os.getenv('DB_NAME')

connection_url = f'postgresql+psycopg2://{DB_USER}:{DB_PASSWORD}@{DB_HOST}:{DB_PORT}/{DB_NAME}'
engine = create_engine(connection_url)

leagues = ['INT-European Championships']
seasons = [2000, 2004, 2008, 2012, 2016, 2020, 2024]

fbref = sd.FBref(leagues=leagues, seasons=seasons)
euro_schedule = fbref.read_schedule()

euro_schedule['date'] = pd.to_datetime(euro_schedule['date'])
euro_schedule['time'] = pd.to_datetime(euro_schedule['time']).dt.time

import hashlib

# CrÃ©er un nouvel index basÃ© sur le hachage de la colonne 'game'
euro_schedule.reset_index(inplace=True)
euro_schedule['index'] = euro_schedule['game'].apply(lambda x: int(hashlib.sha256(x.encode()).hexdigest(), 16))
euro_schedule['home_g'] = euro_schedule['score'].apply(lambda x: x.split('â')[0] if not pd.isna(x) else x)
euro_schedule['away_g'] = euro_schedule['score'].apply(lambda x: x.split('â')[1] if not pd.isna(x) else x)
# int for goals
euro_schedule['home_g'] = pd.to_numeric(euro_schedule['home_g'], errors='coerce')
euro_schedule['away_g'] = pd.to_numeric(euro_schedule['away_g'], errors='coerce')

# DÃ©finir le nouvel index
euro_schedule.set_index('index', inplace=True)

euro_schedule
```

```python
# Display scores if not null and not '-' in it 
def is_score_valid(score):
    if pd.isna(score):
        return False
    return True

euro_schedule['score'].apply(is_score_valid).sum()
```
